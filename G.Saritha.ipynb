```python
#finding the dummie values for non intiger data
dummies=pd.get_dummies(dataset.State)
dummies.head()
```

```python
merge=pd.concat([dataset,dummies],axis=1)
merge.head()
```

```python
#droping the state feature/encoding the catagorical data
dataset=merge.drop(['State','California'],axis=1)
dataset
```

```python
dataset.shape
```

```python
#extracting indipendent variables
x= dataset.drop('Profit', axis =1)
x.head()
```

```python
x.shape
```

```python
#extracting dependent variable
y=dataset['Profit']
y.head()
```

```python
y.shape
```

```python
#spitting the dataset into training set and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2,random_state =1)
x_train.shape,x_test.shape,y_train.shape,y_test.shape

```

```python
#appling linear regrassion on train dataset
from sklearn.linear_model import LinearRegression
mlr = LinearRegression ()
mlr.fit(x_train,y_train)
```

```python
#predicting the values
x_test.head()
```

```python
x_test.shape
```

```python
y_pred =mlr.predict(x_test)
y_pred
```

```python
y_pred.shape
```

```python
y_test
```

```python
print('trian score:',mlr.score(x_train,y_train))
```

```python
print('test score:',mlr.score(x_test,y_test))
```

```python
#comparing actual output(y_test) values with predicted values(y_pred)
dataset1 =pd.DataFrame({'actual':y_test,'predicted':y_test-y_pred})
dataset1.head()
```
